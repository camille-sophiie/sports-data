```python
# Libaries
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Define the new working directory path
new_working_directory = r'/Users/nicol1/Documents/GitHub/sports-data'

# Change the current working directory
os.chdir(new_working_directory)
```

```python
# Define the path 
data_folder = 'data/'

# Load the CSV files to df
cleaned_df = pd.read_csv(data_folder + 'cleaned_df.csv')
cleaned_df
```

```python
# Filter the position for goalkeeper
goalkeeper_df = cleaned_df[cleaned_df['position'] == 'Goalkeeper']
goalkeeper_df
#print(goalkeeper_df.head)
```

```python
# Filter for Defender
defender_df = cleaned_df[cleaned_df['position'] == 'Defender']
defender_df
```

```python
# Filter for Midfield
midfield_df = cleaned_df[cleaned_df['position'] == 'Midfield']
midfield_df
```

```python
# Filter for Striker
striker_df = cleaned_df[cleaned_df['position'] == 'Attack']
striker_df
```

```python
# Scatterplot goalkeeper 

# height
plt.scatter(goalkeeper_df['height'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('Height')
plt.show()

# age
plt.scatter(goalkeeper_df['age'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('Age')
plt.show()

# year
plt.scatter(goalkeeper_df['year'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('number of seasons')
plt.show()

# average minutes
plt.scatter(goalkeeper_df['average_minutes'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('Average Minutes played')
plt.show()

# red cards average
plt.scatter(goalkeeper_df['red_cards_avg'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('Average Red Cards per game')
plt.show()

# yellow cards average
plt.scatter(goalkeeper_df['yellow_cards_avg'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('Average yellow cards per game')
plt.show()

# average assists
plt.scatter(goalkeeper_df['avg_assists_per_year'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('average assists per game')
plt.show()

# average goals
plt.scatter(goalkeeper_df['avg_goals_per_year'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('average goals per game')
plt.show()

# average games per year
plt.scatter(goalkeeper_df['avg_games_per_year'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('average games played per season')
plt.show()

# League
plt.scatter(goalkeeper_df['current_club_domestic_competition_id'], goalkeeper_df['highest_market_value'], c ="pink")
plt.title('League')
plt.show()

# Country
#plt.scatter(goalkeeper_df['country'], goalkeeper_df['highest_market_value'], c ="pink")
#plt.title('Nationalaty')
#plt.show()

# foot
#plt.scatter(goalkeeper_df['foot'], goalkeeper_df['highest_market_value'], c ="pink")
#plt.title('Foot')
#plt.show()
```
